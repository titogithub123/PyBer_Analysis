```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month. 
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot. (Line Chart)
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend. 
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes. 
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("Pybear Fare by Month")
# Add the legend. 
plt.legend()
```

```python
# Recreate (Amend) the plot with color, width and marker updates.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")
# Create labels for the x and y axes. 
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("Pybear Fare by Month")
# Add a grid.
plt.grid()
# Add the legend. 
plt.legend()
```

```python
# Recreate (Amend) Skill Drill
plt.plot(x_axis, y_axis, marker="d", color="magenta", linewidth=1, label="Dallas")
# Create labels for the x and y axes. 
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("Pybear Fare by Month")
# Add a grid.
plt.grid()
# Add the legend. 
plt.legend()
```

```python
# Create a bubble chart using scatter plot as base.
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Use list comprehension to multiply each data point by 3 to make it bigger and easier to distinguish. 
plt.scatter(x_axis, y_axis, s = [i * 10 for i in y_axis])
plt.ylim(0, 50)
```

```python
# Create a scatter plot using the object-oriented approach
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a bublle chart using the object-oriented interface approach.
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Create a bublle chart using the object-oriented interface approach.### Skill-Drill
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c="lime", s= [i * 5 for i in y_axis], 
          alpha=0.3, edgecolors="black", label="Boston", linewidth=2)
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.gca().invert_yaxis()
plt.legend()
```

```python
# Create our first pie chart
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Add percentage and explode "July" to pie chart.
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors =["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct="%.1f%%")
plt.show()
```

```python
# Create a pie chart using object oriented interface
fig, ax = plt.subplots()
ax.pie(y_axis, 
      labels=x_axis,)
plt.show()
```

```python
# Create a pie chart using object oriented interface. ## Skill-Drill
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis, 
       labels=x_axis, 
       autopct="%.1f%%",
       explode=explode_values,
       shadow="True",
       startangle=95,
       counterclock=False,
       colors=colors
      )
plt.show()

```
